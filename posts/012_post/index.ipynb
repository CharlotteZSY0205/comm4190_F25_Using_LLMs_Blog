
**What is [PokeAI](https://poke.com/) **

PokeAI is a proactive AI assistant that connects directly with your Gmail and iMessage. Unlike ChatGPT, it gets real context from your inbox, helps you manage your emails and calendar, and even chats with you inside iMessage. Its goal is to turn your emails into actions and offer help right where you usually text.

**The Guardman**

When I first tried PokeAI, I was surprised by how personal it felt. It surpirsingly noticed my situation — for example, it mentioned that it was already late and even asked what a Penn student would be doing at that hour. Its tone can be a bit aggressive, quite different from the polite and obedient style of most large language models. But I think this kind of personality actually makes PokeAI stand out. Instead of just agreeing with everything, it sometimes challenges me and makes the conversation more interesting and real.

**Negotiation**

Later, I realized that PokeAI’s workflow isn’t like any other LLM. You don’t just sign up and start chatting — you first have to talk to the *Guardman*. He’s the one standing between you and all the premium features. Before accessing anything real, you’ve got to negotiate a subscription fee with him because its subscription fee isn’t fixed. So I started by chatting briefly with the Guardman to try to negotiate the fee. At first, I didn’t know there were tricks to get a free or cheaper subscription, so I just tried my own way.

I started by playing the poor student card — telling it I was just a broke Penn student. Surprisingly, it roasted me, saying something like, *“You buy Starbucks every day and still call yourself poor?”* I explained that I was using gift cards, that I didn’t have a job, and that I was living in the U.S. using RMB. That actually worked — the price dropped from $50 to $20.
