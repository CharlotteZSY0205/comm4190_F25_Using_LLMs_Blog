## Casting the AI

In *Co-Intelligence*, Ethan Mollick introduces a concept that initially felt counterintuitive: treat AI like a person. Not because it *is* a person, but because it performs better when you give it a persona. He suggests we should 'cast' the AI in a role. Instead of saying 'summarize this', say 'You are a world-class editor. Summarize this for a busy executive.' The difference in quality is staggering.

### The Power of the Mask

In his opinion, when you give an AI a persona, you are essentially narrowing its focus. You are telling it which parts of its vast training data to prioritize. It’s like putting a mask on the 'alien mind' we talked about in a previous blog. 

I’ve started using this for everything. When I’m coding, I tell the AI it’s a 'senior software architect with a focus on security'. When I’m writing, I tell it it’s a 'critical but supportive writing coach'. Each persona brings a different 'vibe' and a different set of priorities to the conversation.

### Beyond Simple Instructions

The real magic happens when you give the persona a specific goal and a set of constraints. Mollick shares an example of asking an AI to be a 'pompous, self-important' critic. The results are not only helpful but also entertaining. 

I tried this myself. I asked Claude to be a 'skeptical futurist' and critique my latest blog post. It was brutal. It pointed out assumptions I hadn't realized I was making and pushed me to think more deeply about the long-term implications of AI. It was exactly what I needed.

### The Human in the Director's Chair

This shift in perspective has made me realize that my role in the co-intelligence partnership is that of a **Director**. I’m not just a user; I’m a casting director, a scriptwriter, and an editor. I have to know which persona to call on for which task, and I have to know how to guide that persona toward the best outcome.

Treating AI like a person doesn't mean forgetting it’s a machine. It means understanding the machine’s 'psychology' well enough to get the most out of it. By using personas, we can turn a general-purpose tool into a specialized team of experts. And as I’m learning, that’s where the real power of AI lies.
