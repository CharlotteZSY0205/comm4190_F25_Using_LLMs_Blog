### My prompt for Sora
Through a wide-angle black-and-white security camera recording, we see in the darkness a dog named Pineapple—a 20-pound miniature golden retriever with a big fluffy tail—sneak across the room. The dog quietly climbs onto a piano bench and sits down, bottom squarely on the seat, posture upright, completely focused. There is no music at first.

Suddenly, the dog stretches out both front paws and starts playing an extremely fast, technically demanding piano piece. Its paws move at an incredible speed, leaping across the keys so quickly that the piano begins to smoke. While furiously playing, the dog is also screaming wildly.

Then, without warning, the piano explodes and collapses. The dog panics, falls off the bench, and runs away.
