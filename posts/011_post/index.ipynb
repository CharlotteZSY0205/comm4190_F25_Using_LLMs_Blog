## Possible reasons
**The Mandela Effect**: Many people confidently remember a seahorse emoji. This collective false memory is likely because seahorses are ocean creatures, and numerous other marine animals like tropical fish (🐠), sharks (🦈), and jellyfish already have their own emojis, so it seems logical that a seahorse would, too.

**AI Hallucinations:** Large Language Models (LLMs) like ChatGPT are known to struggle with this question. When asked for the seahorse emoji, they may enter a "recursive failure loop," incorrectly suggesting various fish, a unicorn (🦄), or other animals before often settling on a wrong answer. This happens because the AI tries to find the closest match to "seahorse" in its data but has no correct token to output
