* **Style:** Handheld iPhone shots, cinematic framing (vertical + horizontal mix, for TikTok/IG and YouTube).

🎥 Technical Tips

Angles: Alternate between handheld close-ups (emotional intimacy) and wide establishing shots (Disney’s grandeur).

Framing: Rule of thirds, sometimes center frame for dramatic emphasis.

Camera height: Waist level for casual shots, floor-level upward for powerful shots.

Editing: Sync cuts with beat drops; overlay occasional lyric text in Taylor-style fonts.
