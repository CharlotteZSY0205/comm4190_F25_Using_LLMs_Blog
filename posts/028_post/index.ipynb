I asked ChatGPT to speak as OpenAI CEO Sam Altman and respond to Webster’s 2025 Word of the Year, “Slop“.

## Slop
Merriam-Webster’s human editors have chosen slop as the 2025 Word of the Year. We define slop as “digital content of low quality that is produced usually in quantity by means of artificial intelligence.” All that stuff dumped on our screens, captured in just four letters: the English language came through again.

The flood of slop in 2025 included absurd videos, off-kilter advertising images, cheesy propaganda, fake news that looks pretty real, junky AI-written books, “workslop” reports that waste coworkers’ time… and lots of talking cats. People found it annoying, and people ate it up.
I think “slop” is an uncomfortable word, but it is also an honest one.

When we released products like ChatGPT, DALL·E, and more recently tools that can generate longer texts, code, or images, we knew this would happen. When creation becomes easy and fast, the internet fills up quickly. Some people use these tools to think more clearly or build things they could not build before. Others use them to produce a lot of content with very little intention behind it.

## Sam Altman "says":
I see this every day in our own products. ChatGPT can help someone deeply understand a topic, or it can generate paragraphs that look fine but mean very little. DALL·E can be used to prototype real design ideas, or to churn out endless variations that add no value. The tool is the same. The difference is how it is used.

I do not think this means we should stop building or slow down progress. I think it means we have to take responsibility for what we are amplifying. At OpenAI, a lot of our recent work has been about steering our products toward usefulness instead of volume. That shows up in things like better defaults, more emphasis on reasoning, and features that encourage people to ask better questions rather than just generate more output.

If “slop” is the word of the year, I take it as a signal that the next phase matters more than the first one. The first phase was about proving that these systems work. The next phase is about taste, judgment, and alignment with real human goals. That is true for us, and it is true for the entire industry.

In the long run, the value of AI will not be measured by how much content it can produce, but by how much clearer it helps people think and how much better it helps them decide what is worth making in the first place.
