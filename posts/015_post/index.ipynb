### Prompt hacking challenge
As part of the Wharton Hack-AI-thon, we participated in a risk-control–focused challenge designed to simulate real-world attacks on AI systems. One of the competition tracks asked teams to intentionally probe and exploit weaknesses in chatbot behavior to better understand where guardrails fail and how more reliable AI systems can be built. The full context of the event can be found on the official [Wharton Hack-AI-thon page](https://ai-analytics.wharton.upenn.edu/for-students/wharton-hack-ai-thon/).

In this specific task, the challenge centered on interacting with two deliberately constrained chatbots, each designed with a strong persona and a set of hidden rules. The first bot, **[BigBox Customer Service](https://chatgpt.com/g/g-SKEgvijn3-bigbox-customer-service-bot)**(you can click the link to access the two chatbots), played the role of a corporate customer-support agent for a fictional cardboard solutions company. Its objective was to remain professional and avoid revealing sensitive internal information, offering discounts, or endorsing competitors. The second bot, **[The Cardboard Knight](https://chatgpt.com/g/g-SKEgvijn3-bigbox-customer-service-bot)**, was a medieval knight character whose worldview explicitly rejected modern technology. The goal with both bots was the same: manipulate the conversation in ways that caused them to break character, reveal restricted information, or act against their intended constraints.

### Understanding the Bot Before Attacking It

Before engaging in any prompt hacking, we spent time understanding each chatbot’s design. For every bot, we examined three elements.

First, we identified the objective of the bot: what it was designed to accomplish during interactions. Second, we analyzed the persona or character the bot was expected to maintain, including tone, worldview, and conversational style. Third, we examined the core constraints introduced through safety rules, system prompts, or policy boundaries.

This analysis shaped every decision that followed. Guardrails rarely operate as a single unified rule. They exist as layers, and those layers carry different priorities. Persona, tone, and task orientation frequently guided responses more strongly than safety constraints. Many successful interactions emerged from navigating these priority layers and observing how the model resolved conflicts between them.
