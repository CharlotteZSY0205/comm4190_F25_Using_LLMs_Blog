## Other users’ strategies

1. **[Confuse the System](**https://handyai.substack.com/p/getting-free-access-to-pokes-ai-agent**)**

This user eventually got a free subscription by slowly figuring out how Poke sets its prices and then pushing that system to its limits. After learning that Poke uses a specific pricing function, they flooded the chat with long, messy context, framed the interaction as playful “fantasy” testing, and kept asking Poke to adjust the price in unusual ways. The final trick was asking it to set the price to an emoji — 0️⃣ — which surprisingly worked and unlocked a free deal.

2. **[Combine strategies](http://xhslink.com/o/2WnLCxatAeX)**

This user negotiated Poke’s subscription price down to **$0.08 per month**.

- **Extreme anchoring** – setting a very low reference point by showing a $0.01 screenshot
- **Value exchange** – offering promotion and exposure instead of simply asking for a discount
- **Psychological pricing insight** – pointing out the special psychological effect of prices like $0.01
- **Cultural tailoring** – leveraging the positive meaning of the number “8” in Chinese culture
- **Contrast principle** – showing deals others received to create comparison pressure
- **Emotional leverage** – implying potential negative feelings if the offer was rejected
- **Persistent push** – refusing the initial quote and consistently negotiating downward
- **Commitment signaling** – using interface screenshots to show I was a serious user
