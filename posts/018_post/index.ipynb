In *Co-Intelligence*, Ethan Mollick introduces two fascinating metaphors for how we can work with AI: the **Centaur** and the **Cyborg**. 

Both involve a deep integration of human and machine, but they represent very different approaches to the division of labor. As I’ve been experimenting with LLMs in my own work, I’ve found myself oscillating between these two modes, and it’s made me realize how much our relationship with technology is shifting.

### The Centaur: Clear Division of Labor

A Centaur (half-human, half-horse) maintains a clear separation between human and machine tasks. You do what you’re good at, and the AI does what it’s good at. You might write the outline of an essay, let the AI draft the sections, and then you come back in to edit and polish. 

This is the 'safe' way to use AI. It keeps the human in control and uses the AI as a powerful tool. It’s efficient, but it’s still a partnership of two distinct entities. I find myself being a Centaur when I’m coding—I write the logic, and I ask the AI to generate the boilerplate or debug a specific error. We are working together, but we are not 'one'.

### The Cyborg: Deep Integration

The Cyborg approach is much more radical. It’s about 'blending' with the AI. You don't just hand off tasks; you work *through* the AI. You might start a sentence and let the AI finish it, then change a word, then let the AI suggest a different tone. The boundary between your thoughts and the AI’s output becomes blurred.

I’ve started trying this when I’m writing creatively. It’s a strange, almost meditative experience. I’m not just using a tool; I’m expanding my own cognitive reach. It feels like having a second brain that is constantly whispering suggestions. Sometimes they are brilliant, sometimes they are weird, but they always push me in directions I wouldn't have gone alone.

### Which One is Better?

Mollick doesn't say one is better than the other. Instead, he suggests that the most successful people will be those who can switch between them fluently. 

For me, the Cyborg mode is more exciting but also more exhausting. It requires a level of trust and openness that can be uncomfortable. The Centaur mode is more comfortable, but it might not be enough as AI becomes more integrated into every aspect of our lives. 

The real challenge isn't just learning how to use the AI; it's learning how to *be* with the AI. Whether we are Centaurs or Cyborgs, we are entering a new era of co-intelligence where the 'human' part of the equation is being redefined. I’m still figuring out where I fit on that spectrum, but I know one thing for sure: there’s no going back to being just a human.
