I recently started reading Ethan Mollick’s *Co-Intelligence*, and one concept immediately hit home: **The Jagged Frontier**. 

If you’ve spent any significant time with LLMs, you know exactly what he’s talking about. It’s that invisible, uneven line where AI’s capabilities suddenly drop off a cliff. On one side of the line, the AI can write a complex Python script or summarize a 50-page legal document in seconds. On the other side, it fails at basic arithmetic or hallucinates a historical event that never happened.

### My Encounter with the Edge

Last week, I was using Claude to help me brainstorm some marketing copy for a hypothetical product. It was brilliant—creative, witty, and perfectly on-brand. I was impressed. Then, I asked it to count the number of words in one of the paragraphs it just wrote. 

It failed. Repeatedly. 

This is the Jagged Frontier in action. The AI doesn't have a smooth gradient of ability. It’s not like a human who is generally 'good' or 'bad' at a subject. AI is a patchwork of superhuman intelligence and toddler-level confusion. 

### Why the 'Jaggedness' Matters

Mollick argues that the frontier is jagged because AI doesn't 'know' things the way we do. It predicts the next token. This means it can simulate high-level reasoning without actually possessing the underlying logic. 

For me, this realization was both a relief and a warning. It’s a relief because it explains why I sometimes feel like a genius for using AI and other times like a fool for trusting it. It’s a warning because the frontier is constantly moving. With every update to GPT or Claude, the line shifts. Tasks that were 'outside' the frontier yesterday are 'inside' today.

### Living on the Edge

So, how do we live with this? Mollick suggests we should 'always invite AI to the table' but never leave it alone. We have to be the ones testing the frontier. We have to be the ones who notice when the AI has stepped over the line into hallucination.

I’ve started keeping a 'Jagged Frontier Log'—a mental (and sometimes physical) list of things I’ve found that AI is surprisingly bad at. It’s a way of mapping the territory. But as Mollick warns, the map is not the territory, and the territory is changing fast. 

The Jagged Frontier isn't just a technical limitation; it's a psychological one. It requires us to be constantly vigilant, even when the AI seems most capable. It’s a strange way to work—partnering with an entity that is simultaneously more and less intelligent than you are. But as I’m learning, that’s the essence of Co-Intelligence.
